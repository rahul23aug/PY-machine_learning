PCA Algorithm:
   
    *find a vector that minimises distance to data points 
 
    *Projected points are reduced in dimension but map the or
 
   Mathematically: 1/m *(sum over m features of the data point * data point transpose 
This gives a matrix on n x n dimensions (in this case 3x3) called the correlation matrix sigma.
Find reduced matrix for sigma using svd, svd returns 3 matrices the 1st one is the u-reduced matrix (np.linalg.svd())
Based on k dimensions of choice (2) in this case select first k columns
Calculate vector to project on using
U-reduced transpose .(dot) xdata point
Thus getting a k dimension vector for each data point called z
To get original data back use z .(dot) ureduce

```
import numpy as np, pandas as pd
 
class Standard_Scaler():
  def __init__(self,x):
    self.X = x
    self.xmean = self.X.mean()
    self.xstd =  self.X.std()
  def scaler(self):
    for col in self.X.columns:
      self.X[col] = (self.X[col] - self.X[col].mean()) / self.X[col].std()
    return pd.DataFrame(self.X)
 
  def rescaler(self, X):
    X= pd.DataFrame(X)
    for cols in X.columns:
      X[cols] = X[cols] * self.xstd[cols]
      X[cols] = X[cols] + self.xmean[cols]
    return pd.DataFrame(X)
 
 
class PCA :
  def __init__(self, X,n=1):
    self.X = X
    self.Z = np.array([])
    self.ureduce = np.array([])
    self.n_components = n
    self.m = X.shape[0]
    self.X_approx = np.array([])
 
 
  def transform(self):
    attr_shape = self.X.shape[1]
    sum_vals = np.zeros(attr_shape**2).reshape(attr_shape,attr_shape)
    for row in self.X.index:
      arr =  X.iloc[row].values
      arr = arr.reshape(1,arr.shape[0]) 
      sum_vals = (sum_vals + (arr * arr.T))/self.m
    u,s,v = np.linalg.svd(sum_vals)
    u = pd.DataFrame(u)
    u_prime = u[u.columns[:self.n_components]]
    self.ureduce = u_prime
    #print(u_prime.shape)
    Z = []
    for xi in self.X.index:
      x_arr = self.X.iloc[xi].values
      x_arr = x_arr.reshape(x_arr.shape[0],1)
      Z.append(u_prime.T.dot(x_arr).values)
    Z = np.array(Z)
    Z = Z.reshape(Z.shape[0],self.n_components)
    self.Z = Z
    return pd.DataFrame(Z)
 
  def inverse_transform(self):
    self.X_approx = self.Z.dot(self.ureduce.T)
    return pd.DataFrame(self.X_approx)
  
# usage - pca = PCA(X_data,n_components)
```

```
X=pd.DataFrame([[0,0,0],[1,1,0.5],[2,2,2.2],[3,3,2.9],[4,4,4.2],[5,5,4.9],[6,6,6.2],[7,7,6.2],[8,8,9.2],[9,9,9]])
X
#3-D data
```

```
pca = PCA(X,2)
pca.X
```

```
X_reduced = pca.transform()
X_reduced
#Reduced to 2-D
```

```

# Reconstructed data 
X_approx = pca.inverse_transform()
X_approx
```
